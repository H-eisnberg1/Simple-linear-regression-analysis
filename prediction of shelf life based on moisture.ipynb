```python
import numpy as np
import pandas  as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\OneDrive - Danmarks Tekniske Universitet\Desktop\machine learning\Food scince and technology\Linear_Regression_FoodScience_Q1.csv")
df.head()
```

```python
X = df[["Moisture_Content(%)"]].values.reshape(-1,1)
y = df["Shelf_Life(days)"].values
```

```python
#Traing the model and predictng the shelf life 
model = LinearRegression()
model.fit(X,y)
```

```python
# Coeeficent and slope 
slope = model.coef_
intercept = model.intercept_
# Print the values for each coefficient (if multiple features)
for coeff in slope:
    print(f"{coeff:.2f}")
print(f"{intercept:.2f}")
```

```python
# prediction of new shelf life 
moisture = 18 
predicted_shelf_life = model.predict([[moisture]])[0]
predicted_shelf_life
```

```python
# ploting the values and equation 
plt.scatter(X,y,color = "pink",label = "shelf life and moisture content ")
plt.plot(X,y,color = "b",label = "Relation between shelf life and moisture")
plt.xlabel("Moisture %")
plt.ylabel("Shelf life days")
plt.title("Linear regression for shelf prediction based on moisture content ")
plt.legend()
```
